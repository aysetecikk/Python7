```python
# prompt: kaç farklı meslek ar

# Meslek sayısını bulmak için 'Job Title' sütunundaki benzersiz değerleri sayıyoruz.
unique_jobs = df['job_title'].nunique()
print(f"Veri setinde {unique_jobs} farklı meslek bulunmaktadır.")
```

```python
# prompt: kaç tane kolon var
num_columns = len(df.columns)
print(f"Sütun Sayısı: {num_columns} ")

df.columns
```

```python
# Çalışma yılına göre maaş ortalamaları - salary in usd
salary_by_experience = df.groupby('work_year')['salary_in_usd'].mean()

plt.figure(figsize=(12, 6))
sns.barplot(x=salary_by_experience.index, y=salary_by_experience.values)
plt.xlabel('Çalışma Yılı')
plt.ylabel('Ortalama Maaş (USD)')
plt.title('Çalışma Yılına Göre Ortalama Maaş')
plt.show()
```

```python
# kaç tane meslek kategorisi var -salary in usd
num_job_categories = df['job_category'].unique()
print(f"{num_job_categories} adet meslek kategorisi bulunuyor.")
```

```python
#meslek kategorilerine göre -salary in usd
salary_by_job_category = df.groupby('job_category')['salary_in_usd'].mean()

plt.figure(figsize=(16, 8))
sns.barplot(x=salary_by_job_category.index, y=salary_by_job_category.values)
plt.xlabel('Job Category')
plt.ylabel('Average Salary (USD)')
plt.title('Meslek Kategorilerine Göre Ortalama USD Maaş')
plt.xticks(rotation=45, ha='right')
for index, value in enumerate(salary_by_job_category.values):
    plt.text(index, value + 2000, f'{value:.1f}', ha='center')
plt.tight_layout()
plt.show()
```

```python
#deneyim seviyesine göre maaş dağılımı - salary in usd
salary_by_experience = df.groupby('experience_level')['salary_in_usd'].mean()

plt.figure(figsize=(8, 8))
plt.pie(salary_by_experience, labels=salary_by_experience.index, autopct='%1.1f%%', startangle=90)
plt.title('Deneyim Seviyesine Göre Maaş Dağılımı (USD)')
plt.axis('equal')
plt.show()
```

```python
#yaşanılan yere göre maaşlar - salary in usd
salary_by_location = df.groupby('employee_residence')['salary_in_usd'].mean().sort_values(ascending=False)

plt.figure(figsize=(12, 6))
sns.barplot(x=salary_by_location.index, y=salary_by_location.values)
plt.xlabel('Employee Residence')
plt.ylabel('Average Salary (USD)')
plt.title('Average Salary by Employee Residence')
plt.xticks(rotation=90, fontsize=6)
plt.tight_layout()
plt.show()

N = 10
print(f"Top {N} locations with the highest average salaries:")
print(salary_by_location.head(N))
```
